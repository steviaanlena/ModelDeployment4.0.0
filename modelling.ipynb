# **Split Data**

Pada split data ini, kita akan membagi dataset kita menjadi 3, yaitu 70% train, 15% test, dan 15% validation menggunakan stratified sampling dan random state  0.

```python
from sklearn.model_selection import train_test_split

x = df.drop(columns=["churn", "id", "Surname"])
y = df["churn"]

#Split the data into training (70%) and a temporary set (30%)
x_train, x_temp, y_train, y_temp = train_test_split(x, y, test_size=0.3, random_state=0, stratify=y)

#Split the temporary set into validation (50%) and test (50%)
x_val, x_test, y_val, y_test = train_test_split(x_temp, y_temp, test_size=0.5, random_state=0, stratify=y_temp)
```

# **Data Handling**

Pada data handling, dilakukan pengisian na values di kolom CreditCard, yang dilakukan dengan cara check outlier. Apabila ada outlier, maka isi na menggunakan median, dan jika tidak ada, isi na menggunakan mean. Saya juga membuat versi Quadrant formulanya dikarenakan nantinya akan lebih mudah dan cocok untuk dipindahkan ke OOP.

```python
df.info()
```

```python
#Check for outliers using a box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x=x_train['CreditScore'])
plt.title('Boxplot of CreditScore')
plt.xlabel('CreditScore')
plt.show()
```

```python
#Calculate the first quartile (Q1) and third quartile (Q3)
Q1 = x_train['CreditScore'].quantile(0.25)
Q3 = x_train['CreditScore'].quantile(0.75)

#Calculate the interquartile range (IQR)
IQR = Q3 - Q1

#Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

#Check for outliers
outliers = x_train[(x_train['CreditScore'] < lower_bound) | (x_train['CreditScore'] > upper_bound)]
print("Outliers detected using the IQR method:")
print(len(outliers))
```

```python
if len(outliers)>0:
  median_credit_score= x_train['CreditScore'].median()
  x_train['CreditScore'].fillna(median_credit_score, inplace=True)
  x_test['CreditScore'].fillna(median_credit_score, inplace=True)
  x_val['CreditScore'].fillna(median_credit_score, inplace=True)
else:
  mean_credit_score = x_train['CreditScore'].mean()
  x_train['CreditScore'].fillna(mean_credit_score, inplace=True)
  x_test['CreditScore'].fillna(mean_credit_score, inplace=True)
  x_val['CreditScore'].fillna(mean_credit_score, inplace=True)
```

# **Model 1 : Random Forest**

Setelah semua proses data selesai, kita lakukan modelling. Model pertama yang diminta adalah Random Forest. Di sini saya menggunakan RF Classifier karena data targetnya kategorikal, kemudian saya tuning menggunakan grid search dengan parameter criterion dan max depth yang dituning.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

rf = RandomForestClassifier()
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 4, 5]
}

grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, scoring='accuracy')

grid_search.fit(x_train, y_train)
print("Best params:", grid_search.best_params_)
```

```python
best_rf_model = grid_search.best_estimator_
y_val_pred = best_rf_model.predict(x_val)
print("Classification Report for Validation Data:")
print(classification_report(y_val, y_val_pred))
```

```python
y_test_pred = best_rf_model.predict(x_test)
print("Classification Report for Test Data:")
print(classification_report(y_test, y_test_pred))
```

Ternyata, best parameternya adalah gini dengan max depth 5. Setelah dilakukan validation dan testing, bisa dilihat data kita cukup baik, dengan akurasi validasi 86% dan test 85%.

# **Model 2 : XGBoost**

Di model ke 2 ini, saya menggunakan XGBoost sesuai soal. Di sini saya juga tuning lagi parameternya dengan pilihan parameter learning rate, max depth, dan n estimators.

```python
from xgboost import XGBClassifier

xgb = XGBClassifier()

param_grid = {
    'learning_rate': [0.1, 0.01, 0.001],
    'max_depth': [3, 4, 5],
    'n_estimators': [100, 200, 300]
}

grid_search_xgb = GridSearchCV(estimator=xgb, param_grid=param_grid, cv=3, scoring='accuracy')

grid_search_xgb.fit(x_train, y_train)
print("Best params:", grid_search_xgb.best_params_)
```

```python
best_xgb_model = grid_search_xgb.best_estimator_
y_val_pred_xgb = best_xgb_model.predict(x_val)
print("Classification Report for Validation Data:")
print(classification_report(y_val, y_val_pred_xgb))
```

```python
y_test_pred_xgb = best_xgb_model.predict(x_test)
print("Classification Report for Test Data:")
print(classification_report(y_test, y_test_pred_xgb))
```

Ternyata ditemukan akurasinya untuk validation 87% dan 86% untuk testingnya.
